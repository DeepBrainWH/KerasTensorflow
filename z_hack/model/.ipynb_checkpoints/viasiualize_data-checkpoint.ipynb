```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
file1 = "../data/part-r-2009"
file2 = "../data/part-r-2010"
file3 = "../data/part-r-2011"
file4 = "../data/part-r-2012"
file5 = "../data/part-r-2013"

data1 = pd.read_csv(file1, sep='\t', header=None)
data2 = pd.read_csv(file2, sep='\t', header=None)
data3 = pd.read_csv(file3, sep='\t', header=None)
data4 = pd.read_csv(file4, sep='\t', header=None)
data5 = pd.read_csv(file5, sep='\t', header=None)
```

```python
x1 = data1[0].values
x2 = data2[0].values
x3 = data3[0].values
x4 = data4[0].values
x5 = data5[0].values
```

```python
y1 = data1[3].values
y2 = data2[3].values
y3 = data3[3].values
y4 = data4[3].values
y5 = data5[3].values
```

```python
from datetime import datetime
xs1 = [datetime.strptime(d, '%Y-%m-%d').date() for d in x1]
xs2 = [datetime.strptime(d, '%Y-%m-%d').date() for d in x2]
xs3 = [datetime.strptime(d, '%Y-%m-%d').date() for d in x3]
xs4 = [datetime.strptime(d, '%Y-%m-%d').date() for d in x4]
xs5 = [datetime.strptime(d, '%Y-%m-%d').date() for d in x5]
plt.figure(figsize=(20, 10))
plt.plot(xs1, y1, label='2009')
plt.plot(xs2, y2, label='2010')
plt.plot(xs3, y3, label='2011')
plt.plot(xs4, y4, label='2012')
plt.plot(xs5, y5, label='2013')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(25, 60))
ax1 = plt.subplot(5, 1, 1)
ax1.plot(xs1, y1, label='2009')
plt.ylim([2500, 8000])
plt.legend()

ax2 = plt.subplot(5, 1, 2)
ax2.plot(xs2, y2, label='2010')
plt.ylim([2500, 8000])
plt.legend()

ax3 = plt.subplot(5, 1, 3)
ax3.plot(xs3, y3, label='2011')
plt.ylim([2500, 8000])
plt.legend()

ax4 = plt.subplot(5, 1, 4)
ax4.plot(xs4, y4, label='2012')
plt.ylim([2500, 8000])
plt.legend()

ax5 = plt.subplot(5, 1, 5)
ax5.plot(xs5, y5, label='2013')
plt.ylim([2500, 8000])
plt.legend()

plt.show()
```

```python
xx1 = data1[2].values
xx2 = data2[2].values
xx3 = data3[2].values
xx4 = data4[2].values
xx5 = data5[2].values

plt.figure(figsize=(25, 12))
ax1 = plt.subplot(2, 3, 1)
ax1.plot(xs1, xx1, label='2009')
plt.ylim([10000, 2000000])
plt.legend()

ax2 = plt.subplot(2, 3, 2)
ax2.plot(xs2, xx2, label='2010')
plt.ylim([10000, 2000000])
plt.legend()

ax3 = plt.subplot(2, 3, 3)
ax3.plot(xs3, xx3, label='2011')
plt.ylim([10000, 2000000])
plt.legend()

ax4 = plt.subplot(2, 3, 4)
ax4.plot(xs4, xx4, label='2012')
plt.ylim([10000, 2000000])
plt.legend()

ax5 = plt.subplot(2, 3, 5)
ax5.plot(xs5, xx5, label='2013')
plt.ylim([10000, 2000000])
plt.legend()

plt.show()
```
